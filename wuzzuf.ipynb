```python
jobs=[]
for i in range(1,191):
    url=f"https://wuzzuf.net/search/jobs/?a=hpb&q=illustrator&start={i}"
    response=requests.get(url)
    response=response.content
    soup=BeautifulSoup(response,'html.parser')
    div_container=soup.find('div')
    
    containers=div_container.find_all('div',class_='css-1gatmva e1v1l3u10')

    for container in containers:
        name_job=container.find('h2',class_='css-m604qf').text.strip()
        company_name=container.find('a',attrs={'target':'_blank','class':'css-17s97q8'}).get_text()
        company_adress=container.find('span',class_='css-5wys0k').text.strip()
        time_job=container.find('span',class_='css-1ve4b75 eoyjyou0').text
        link=container.find('a',attrs={'target':'_blank','class':'css-o171kl'}).get('href')
        description=container.find('div',attrs={'class':'css-y4udm8'}).get_text()[9:]
    
        jobs.append([name_job,company_name,company_adress,time_job,link,description])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(jobs,columns=['name_job','company_name','company_adress','time_job','link','description'])
```
